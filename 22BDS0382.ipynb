## Univariate analysis

### Subtask:
Analyze each variable individually, including visualizations like histograms, box plots, or frequency tables to understand their distributions.


**Reasoning**:

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['age', 'economic.cond.national', 'economic.cond.household', 'Blair', 'Hague', 'Kennedy', 'Europe', 'political.knowledge']
categorical_cols = ['vote', 'gender']

# Histograms for numerical columns
df[numerical_cols].hist(figsize=(15, 10))
plt.tight_layout()
plt.suptitle('Histograms of Numerical Columns', y=1.02)
plt.show()

# Box plots for numerical columns
plt.figure(figsize=(15, 10))
sns.boxplot(data=df[numerical_cols])
plt.title('Box Plots of Numerical Columns')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Frequency tables and count plots for categorical columns
for col in categorical_cols:
    print(f"\nFrequency Table for {col}:\n")
    display(df[col].value_counts())
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=col)
    plt.title(f'Count Plot of {col}')
    plt.show()
```

# Task
Perform hierarchical clustering on the dataset.

## Select features

### Subtask:
Choose the numerical features from the dataset that will be used for clustering.


**Reasoning**:
Select the numerical columns from the DataFrame and store them in a new DataFrame for clustering.



```python
numerical_cols_for_clustering = ['age', 'economic.cond.national', 'economic.cond.household', 'Blair', 'Hague', 'Kennedy', 'Europe', 'political.knowledge']
df_numerical = df[numerical_cols_for_clustering]
display(df_numerical.head())
```

## Scale data

### Subtask:
Scale the selected numerical features to ensure that all features contribute equally to the distance calculations.


**Reasoning**:
Scale the selected numerical features using StandardScaler as requested.



```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numerical)
```

## Calculate distance matrix

### Subtask:
Compute the distance matrix between data points using an appropriate distance metric.


**Reasoning**:
Compute the distance matrix between data points using an appropriate distance metric.



```python
from scipy.spatial.distance import pdist

distance_matrix = pdist(df_scaled, metric='euclidean')
```

## Perform hierarchical clustering

### Subtask:
Apply hierarchical clustering to the distance matrix using a chosen linkage method.


**Reasoning**:
Apply the linkage function to the distance matrix using the 'ward' method to perform hierarchical clustering.



```python
from scipy.cluster.hierarchy import linkage

linked_matrix = linkage(distance_matrix, 'ward')
```

## Visualize dendrogram

### Subtask:
Plot the dendrogram to visualize the hierarchical structure of the clusters and help in determining the optimal number of clusters.


**Reasoning**:
Plot the dendrogram using the linked matrix to visualize the hierarchical structure and aid in determining the number of clusters.



```python
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(15, 7))
dendrogram(linked_matrix, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

## Visualize clustered data (optional)

### Subtask:
Visualize the clustered data, perhaps using a heatmap or by plotting clusters on a dimensionality reduced representation of the data.


**Reasoning**:
Cut the dendrogram to form clusters based on the chosen number of clusters (visually inspecting the dendrogram suggests 3 clusters), add the cluster labels to the original DataFrame, perform PCA for dimensionality reduction, and create a scatter plot of the first two principal components colored by cluster labels.



```python
from scipy.cluster.hierarchy import fcluster
from sklearn.decomposition import PCA

# Determine the optimal number of clusters by visually inspecting the dendrogram.
# Based on visual inspection of the dendrogram, 3 clusters seem reasonable (longest vertical lines before crossing horizontal lines).
n_clusters = 3

# Cut the dendrogram to form clusters
cluster_labels = fcluster(linked_matrix, n_clusters, criterion='maxclust')

# Add the cluster labels to the original DataFrame
df['cluster'] = cluster_labels

# Perform PCA for dimensionality reduction
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_scaled)
pca_df = pd.DataFrame(data = principal_components, columns = ['principal_component_1', 'principal_component_2'])

# Add cluster labels to the PCA DataFrame
pca_df['cluster'] = cluster_labels

# Create a scatter plot of the first two principal components, coloring by cluster labels
plt.figure(figsize=(10, 8))
sns.scatterplot(data=pca_df, x='principal_component_1', y='principal_component_2', hue='cluster', palette='viridis', legend='full')
plt.title('PCA of Clustered Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```
