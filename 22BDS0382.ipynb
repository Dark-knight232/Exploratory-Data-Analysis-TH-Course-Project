## Univariate analysis

### Subtask:
Analyze each variable individually, including visualizations like histograms, box plots, or frequency tables to understand their distributions.


**Reasoning**:
Generate visualizations for numerical and categorical columns to understand their distributions and display frequency tables for categorical columns.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['age', 'economic.cond.national', 'economic.cond.household', 'Blair', 'Hague', 'Kennedy', 'Europe', 'political.knowledge']
categorical_cols = ['vote', 'gender']

# Histograms for numerical columns
df[numerical_cols].hist(figsize=(15, 10))
plt.tight_layout()
plt.suptitle('Histograms of Numerical Columns', y=1.02)
plt.show()

# Box plots for numerical columns
plt.figure(figsize=(15, 10))
sns.boxplot(data=df[numerical_cols])
plt.title('Box Plots of Numerical Columns')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Frequency tables and count plots for categorical columns
for col in categorical_cols:
    print(f"\nFrequency Table for {col}:\n")
    display(df[col].value_counts())
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=col)
    plt.title(f'Count Plot of {col}')
    plt.show()
```

## Bivariate analysis

### Subtask:
Explore relationships between pairs of variables using techniques such as scatter plots, correlation matrices, or cross-tabulations.


**Reasoning**:
Calculate and display the correlation matrix for the numerical columns, generate scatter plots for interesting pairs, create cross-tabulations for categorical variables, and generate box plots for a categorical and a numerical variable.



```python
# Calculate and display the correlation matrix for numerical columns
numerical_cols = ['age', 'economic.cond.national', 'economic.cond.household', 'Blair', 'Hague', 'Kennedy', 'Europe', 'political.knowledge']
correlation_matrix = df[numerical_cols].corr()
display(correlation_matrix)

# Generate scatter plots for pairs of numerical variables
# Based on the correlation matrix and domain knowledge, let's plot age vs political.knowledge and Blair vs Hague
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='age', y='political.knowledge')
plt.title('Scatter Plot of Age vs. Political Knowledge')

plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='Blair', y='Hague')
plt.title('Scatter Plot of Blair vs. Hague Ratings')

plt.tight_layout()
plt.show()

# Create cross-tabulations for pairs of categorical variables
cross_tab_vote_gender = pd.crosstab(df['vote'], df['gender'])
display(cross_tab_vote_gender)

# Visualize the relationship between a categorical variable and a numerical variable (age by vote)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='vote', y='age')
plt.title('Box Plot of Age by Vote')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
